# only IP formation for unsupervised learning

```python
ip=iris.drop(["CLASS"], axis =1)
```

```python
from sklearn import cluster
km = cluster.KMeans(n_clusters = 3)
km.fit(ip)
```

```python
k=km.predict(ip)
print(k)
```

```python
plt.figure(figsize= (12,5))
plt.scatter(iris.SL, iris.PL, c = k, s= 50, cmap="viridis")
plt.show()
# s = size of the data points
#cmap = collection of gradient-colours to a plot
```

```python
centroid = km.cluster_centers_
print(centroid)
```

# Random dataset from Scikit Learn

```python
from sklearn.datasets import  make_blobs
x,y = make_blobs(n_samples= 300, centers = 4, cluster_std=0.6, random_state = 0)
data=pd.DataFrame(x, y)
```

```python
plt.scatter(x[:,0], x[:,1], s=50)
plt.show()
```

```python
from sklearn import cluster
km2 = cluster.KMeans(n_clusters = 4)
km2.fit(x)
```

```python
yp = km2.predict(x)
data['predict'] = yp
```

```python
data
```

```python
centroid = km2.cluster_centers_
print(centroid)
```

```python
plt.scatter(x[:,0], x[:,1], c=yp, s=50,  cmap="viridis")
plt.scatter(centroid[:,0], centroid[:,1], c = "black", s = 200)
plt.show()
# s = size of the data points
#cmap = collection of gradient-colours to a plot
```

```python
def ClusterIndicesNumpy (clustNum, labels_array):
    return np.where(labels_array==clustNum)
ClusterIndicesNumpy(2, (km2.labels_))
```
